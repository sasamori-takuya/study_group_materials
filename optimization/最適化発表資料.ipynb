## 不等式制約つき最適化問題の解法「ペナルティ法」

「目的関数」を最適化するのではなく，「目的関数+ペナルティ関数」を最適化することを考える  
すると元々の「目的関数」に対する制約つき最適化問題が，
「目的関数+ペナルティ関数」に対する無制約最適化問題に化ける

image.png

ペナルティ法のアルゴリズム（多分）  
0.アルゴリズム停止パラメータ $\epsilon >0$ を固定する．またペナルティパラメータ更新律 $c > 1$ を固定する．  
1.初期点 $x_0 \in \mathbb{R}^n$ を選ぶ．初期ペナルティパラメータ $\rho_0 >0$ を選ぶ．  
2.$\phi_\rho$ の無制約最適化問題のアルゴリズムを解いて $x_1$ を選ぶ（アルミホ+バックトラック等々）．  
3.$|\nabla f(x_1)| < \epsilon$ を満たせばアルゴリズムを停止．そうでなければ続行．  
4.$\rho_1 = c \rho_0$ とする．  
5.$x_1$ と $\rho_1$ に対して手順2を行い，$x_2$ を選ぶ．  
（以下繰り返し）  

$\rho$ が十分小さい → 元々の目的関数の無制約最適化問題とほぼ等しい．解きやすいが求めた点が制約を満たす領域（実行可能領域）にない可能性が高い．  
$\rho$ が十分大きい → 解きづらいが，求めた点は実行可能領域に十分近いところにある  
  
よってペナルティ法は「実行可能領域の外側から徐々に近似していき，元々の制約つき最適化問題の停留点を求めていく」手法と言える

## 不等式制約つき最適化問題の解法「障壁法」

実行可能領域の境界で発散して，実行可能領域の内部で有限値をとるバリア関数を用いて，元々の「目的関数」の制約つき最適化問題を，「目的関数+バリア関数」の無制約最適化問題として考える．

image.png


障壁法のアルゴリズム（多分）  
0.アルゴリズム停止パラメータ $\epsilon >0$ を固定する．またバリアパラメータ更新律 $c < 1$ を固定する．  
1.初期点 $x_0 \in \mathbb{R}^n$ を選ぶ．初期バリアパラメータ $\nu_0 >0$ を選ぶ．  
2.$\psi_\nu$ の無制約最適化問題のアルゴリズムを解いて $x_1$ を選ぶ（アルミホ+バックトラック等々）．  
3.$|\nabla f(x_1)| < \epsilon$ を満たせばアルゴリズムを停止．そうでなければ続行．  
4.$\nu_1 = c \nu_0$ とする．  
5.$x_1$ と $\nu_1$ に対して手順2を行い，$x_2$ を選ぶ．  
（以下繰り返し）  

障壁法を利用する際には，等式制約があってはいけない（いろいろ調べてみたが，不等式制約のみを仮定しているものしかなかった）

注意：等式制約は不等式制約に置き換えることができる．  
$h_1(x) = 0$ は $h_1(x) \geq 0$ かつ $h_1(x) \leq 0$ であるので，
$h_1(x) \leq 0, -h_1(x) \leq 0$ とすれば，等式制約をすべて不等式制約の形に直せる．

$\nu$ が十分大きい → 求めた点は，実行可能領域の中心部分近くにある．ただし往々にして元々の目的関数の制約つき最適化問題の局所最適解（もどき）は実行可能領域の境界部分にある  
$\nu$ が十分小さい → 求めた点は，実行可能領域の境界近くにあるし，必ず実行可能領域に属する．バリア関数部分が発散しやすいので求めづらいが，「$f(x)+\infty$」の最小化問題のようなものなので，求めた点は$f$の局所最適解もどきに非常に近い．  
  
よって障壁法は「実行可能領域の内側から徐々に近似していき，元々の制約つき最適化問題の停留点を求めていく」手法と言える

## 不等式制約つき最適化問題の解法「主双対内点法」

不等式制約つき最小化問題
\begin{align*}
  \begin{cases}
    \mathrm{Minimize \quad} f(\boldsymbol{x}) \\
    \mathrm{subject \,\, to \quad} g_i(\boldsymbol{x}) \leq 0, h_l(\boldsymbol{x}) = 0 \quad (i=1,\ldots,m,\,\,l=1,\ldots,r)
  \end{cases} \tag{2.57}
\end{align*}
を考える．

* 元々は線形計画問題に対するアルゴリズム
* 非線型計画問題に持ち込み，内点法の難点を克服する研究が進んだ
* 逐次2次計画法と並んで，非線型計画問題の主要な解法のひとつ

### 内点法

実行可能領域内の点において，すべての不等式制約条件で等号成立しない点のことを **内点** という．  
実行可能領域が $S$ であり，不等式制約が $g_1,\ldots, g_m$ の$m$個の関数があるとしたら
$\{x \in S \mid I_{-}(x) = \{1,\ldots,m\}\}$ に属する点のことを指す．

ひとつの内点が与えられているとき，その内点に十分近い近傍では不等式制約が成り立っているはずなので，不等式制約を考慮せずに等式制約のみに注意しながら目的関数の減少方向を逐次的に探していく方法を「内点法」という．  
内点法においては「逐次更新の際に境界に近づきすぎない」ように考慮する．  
内点法と対になるのが単体法であり，こちらは逆に境界上を更新していくアルゴリズムである（境界から単体が作られるので，そこのみで考えるということ）．  
一般に境界上での更新は効率が悪いことが多いので，内点法はこれを回避している（小規模や中規模ならば単体法のほうが良いらしいが，大規模だと内点法に軍配が上がるらしい）．

また内点法は理論的に収束が保障されていることが多いらしい（詳細は不明）．

内点列を逐次的に構成していく方法はいくつかある．  
まずひとつ目は，実行可能領域の内部で定義され，実行可能領域の境界に近づくと発散するような補正を加えて目的関数の最小化を目指す方法である．  
これにうってつけなのがバリア関数である．  
目的関数にバリア関数（with重み）を加えた「拡張された目的関数」を定義して，「重みを固定して次の内点を探索→次の内点を見つけたらそれを固定して重みを探索→...」を交互に繰り返していく．  
  
バリア関数の亜種としてポテンシャル関数を考えるポテンシャル減少法というものもある．  
最適解以外の境界では$+\infty$をとり，最適解では$-\infty$をとるような関数（ポテンシャル関数）を導入して，ポテンシャル関数値が十分小さくなるまで更新を繰り返していく．  
  
別の方法として，「与えられた内点を $(1,\ldots,1)$ という点にアフィン変換→アフィン変換後の実行可能領域の内部内の球面上で目的関数を最小化させる点を探索→その探索結果の点を逆アフィン変換」を繰り返していくアフィンスケーリング法というものもある．  
アフィン変換を用いることで，内点が境界付近にあったとしても一旦アフィン変換して中心に寄せることによって計算の困難を回避させるという効果がある．  
  
アフィン変換の代わりに射影変換を用いる射影変換法（またはKarmarkar法）もある．  
  
「境界に近づかない」ために「境界から最も離れた点を集めた集合」に着目する方法もある．  
その集合のことは「中心パス」と呼ばれ，その中心パスから離れすぎないという制約のことで内点の逐次更新を行う方法をパス追跡法という．  
  


### スラック変数

等式制約を不等式制約に置換するときには「=」を「$\leq$ and $\geq$」置き換えるだけだったが，不等式制約を等式制約に置換するには補助変数（**スラック変数**と呼ばれる）を用いる必要がある．

たとえば $g(x) \leq 0$ は $g(x) + z = 0, z \geq 0$ とすればよい．  
この等式制約を満たす領域を $D'_x \times D'_z$ とする．  
元々の不等式制約を満たす領域（つまり実行可能領域）を$D$ とする．  
$(x, z) \in D'_x \times D'_z$ であるならば
$0 \leq z = -g(x)$ なので $g(x) \geq 0$ となる．すなわち $x \in D$ である．  
つまり $D'_x \subset D$ である．  
逆の場合，つまり $x \in D$ の場合は単に $(x, -g(x)) \in D'_x \times D'_z$ である(不等式制約から $-g(x) \geq 0$ は明らか)．  
  
つまりスラック変数を用いて不等式制約を等式制約に変えたとき，実行可能領域は本質的には変化しない．  
新たな不等式制約として $z \geq 0$ が増えるが，等式制約と不等式制約が混ざったような制約に関して，不等式制約を等式制約に置換することで，最適化を行う上で都合がよくなることが多い．  

(2.57)をスラック変数を用いて変化させると，
\begin{align*}
  \begin{cases}
    \mathrm{Minimize \quad} f(\boldsymbol{x}) \\
    \mathrm{subject \,\, to \quad} g_i(\boldsymbol{x}) + z_i = 0, h_l(\boldsymbol{x}) = 0, z_i \geq 0 \quad (i=1,\ldots,m,\,\,l=1,\ldots,r)
  \end{cases}
\end{align*}
となる．


ここで
\begin{align*}
  &\tilde{f}(x_1,\ldots,x_n,z_1,\ldots,z_n) := f(x_1,\ldots,x_n) \\
  &\tilde{h}_l(x_1,\ldots,x_n,z_1,\ldots,z_n) := h_l(x_1,\ldots,x_n) \quad (l=1,\ldots,r) \\
  &\tilde{h}_{r+i}(x_1,\ldots,x_n,z_1,\ldots,z_n) := g_i(x_1,\ldots,x_n) + z_i \quad (i=1,\ldots,m)
\end{align*}
と置くと，上述の最適化問題は
\begin{align*}
  \begin{cases}
    \mathrm{Minimize \quad} \tilde{f}(x_1,\ldots,x_n,z_1,\ldots,z_n) \\
    \mathrm{subject \,\, to \quad} \tilde{h}_l(x_1,\ldots,x_n,z_1,\ldots,z_n) = 0, z_i \geq 0 \quad (i=1,\ldots,m,\,\,l=1,\ldots,r+m)
  \end{cases}
\end{align*}
と表すことができる．

以降は簡単のために
\begin{align*}
  \begin{cases}
    \mathrm{Minimize \quad} f(\boldsymbol{x}) \\
    \mathrm{subject \,\, to \quad} h_l(\boldsymbol{x}) = 0 \quad (l=1,\ldots,r), \boldsymbol{z} \geq \mathbf{0} 
  \end{cases} \tag{2.74}
\end{align*}
という形式の問題を考える．

**注意**  
$x = (x_1,\ldots, x_n) \in \mathbb{R}^n$ に対して $x \geq 0$ であるとは，$x_i \geq 0$ ($i=1,\ldots,n$) が成り立つことである．

(2.74)の非負制約 $x_j \geq 0$ に対して $\nu>0$ をバリアパラメータとして、対数バリア関数 $-\nu \log x_j$ を考えて，
\begin{align*}
  \begin{cases}
    \mathrm{Minimize \quad} f(\boldsymbol{x}) - \nu \displaystyle{\sum_{j=1}^n} \log x_j \\
    \mathrm{subject \,\, to \quad} h_l(\boldsymbol{x}) = 0 \quad (l=1,\ldots,r),\quad g_k(\boldsymbol{x}) < 0 \quad (k=1,\ldots,n)
  \end{cases} \tag{2.74}
\end{align*}
という不等式制約つき最適化問題を考える．  
ただしここで
\begin{align*}
  g_k(\boldsymbol{x}) := - x_k
\end{align*}
である．  
つまり不等式制約は $x_k > 0$ ($k=1,\ldots,n$) である．  
  
この不等式制約問題に関するKKT 条件について考える．  
まずLagrange関数の停留条件は
\begin{align*}
  \boldsymbol{0} 
  &= \nabla f(\boldsymbol{x}) -  \nabla \left(\nu \sum_{j=1}^n \log x_j\right)
   + \sum_{k=1}^n \kappa_k \nabla g_k(\boldsymbol{x})
   + \sum_{l=1}^r \mu_l \nabla h_l(\boldsymbol{x})\\
  &= \nabla f(\boldsymbol{x}) - \left(\dfrac{\nu}{x_1},\ldots,\dfrac{\nu}{x_n}\right)^\top
  - \left( \kappa_1,\ldots, \kappa_n\right)^\top
   + \sum_{l=1}^r \mu_l \nabla h_l(\boldsymbol{x})\\
  &= \nabla f(\boldsymbol{x}) - \left(\lambda_1,\ldots,\lambda_n\right)^\top
  - \left( \kappa_1,\ldots, \kappa_n\right)^\top
   + \sum_{l=1}^r \mu_l \nabla h_l(\boldsymbol{x})\\
  &= \nabla f(\boldsymbol{x}) - \boldsymbol{\lambda}
  - \boldsymbol{\kappa}
   + J \boldsymbol{h}(\boldsymbol{x})^\top \boldsymbol{\mu}
\end{align*}
である．ただしここで
\begin{align*}
  \lambda_j := \dfrac{\nu}{x_j} \in \mathbb{R},\quad
  J \boldsymbol{h}(\boldsymbol{x})
  := \left( \dfrac{\partial h_i}{\partial x_j}(\boldsymbol{x})\right)_{1 \leq i \leq r, 1 \leq j \leq n} \in \mathbb{R}^{r \times n},\quad \boldsymbol{\mu} := (\mu_1,\ldots,\mu_r)^\top \in \mathbb{R}^r
\end{align*}
である(不等式制約より $x_j \neq 0$ なので $\lambda_j$ は定義可能であり，特に $x_j > 0$ かつ $\nu>0$ であるから $\lambda_j>0$)．

次に相補性条件について考えると，
\begin{align*}
  g_k(\boldsymbol{x}) < 0, \kappa_k \geq 0, \kappa_k g_k(\boldsymbol{x}) = 0 \quad (k=1,\ldots, n)
\end{align*}
であるが $g_k(\boldsymbol{x}) \neq 0$ であるので $\kappa_1=\cdots=\kappa_n=0$ でなければならない．

以上より，元々の制約条件を加えると，KKT条件は
\begin{align*}
  &\nabla f(\boldsymbol{x}) - \boldsymbol{\lambda}
   + J \boldsymbol{h}(\boldsymbol{x})^\top \boldsymbol{\mu}
  = \boldsymbol{0} \\
  & x_j > 0 \quad (j=1,\ldots,n)\\
  & h_l(\boldsymbol{x}) = 0 \quad (l=1,\ldots,r) \\
  & \nu = \lambda_j x_j,\quad \lambda_j > 0 \quad (j=1,\ldots,n)
\end{align*}
である．

## 不等式制約つき最適化問題の解法「（拡張Lagrange）乗数法」

一言で標語的に表すならば「Lagrange乗数法+ペナルティ法」

等式制約問題
\begin{align*}
  &\mathrm{Minimize \quad} f(\boldsymbol{x})\\
  &\mathrm{subject \, to \quad} h_l(\boldsymbol{x}) \quad (l=1,\ldots, r) \tag{2.48}
\end{align*}
を考える．

次の関数を考える：
\begin{align*}
  L^a_\sigma(\boldsymbol{x}, \boldsymbol{\mu})
  &:=
   f(\boldsymbol{x}) + \boldsymbol{\mu}^\top \boldsymbol{h}(\boldsymbol{x}) + \dfrac{\sigma}{2} \|\boldsymbol{h}(\boldsymbol{x})\|^2 \\
  &= f(\boldsymbol{x}) + \sum_{l=1}^r \mu_l h_l(\boldsymbol{x}) + \underbrace{\dfrac{\sigma}{2} \sum_{l=1}^r h_l(\boldsymbol{x})^2} \tag{2.78}
\end{align*}

右辺の第3項目は等式制約に関するペナルティ関数である．  
$\sigma > 0$ はペナルティパラメータであり，$\sigma \to \infty$ にしていくほどペナルティが強くなる．
なお，$\sigma = 0$ のとき，この関数は Lagrange 関数と同じである．  
よってこの関数は **拡張Lagrange関数** と呼ばれる．  
$\boldsymbol{x}$ について勾配をとると
\begin{align*}
  \nabla_\boldsymbol{x} L^a_\sigma(\boldsymbol{x}, \boldsymbol{\mu})
  = \nabla f(\boldsymbol{x}) + J \boldsymbol{h}(\boldsymbol{x})^\top (\boldsymbol{\mu} + \sigma \boldsymbol{h}(\boldsymbol{x}))
\end{align*}
である．

よって $\boldsymbol{\mu}_k$ の更新式としては $\boldsymbol{\mu}_{k+1} = \boldsymbol{\mu}_k + \sigma \boldsymbol{h}(\boldsymbol{x})$ が妥当．

Lagrange関数の未定乗数法のデメリットとしては「Lagrange関数の停留点しか求められない（正確に言えば，求めた停留点が極点なのか鞍点なのかわからない）」点である．  
一方，拡張Lagrange関数を用いるとペナルティ関数の部分が凸になっているため，最適解を求められる可能性が高くなる．

image.png

**命題2.11の意味**  
ある程度の仮定の下で，拡張Lagrange関数を用いた最適化は，「有限の」ペナルティパラメータで元々の目的関数の制約つき最適化問題の最適解が得られる．

ペナルティ法はペナルティパラメータを無限に飛ばさないと，元々の最適解は得られない．  
しかし拡張Lagrange関数による乗数法ではある程度の大きさをもつ有限なペナルティパラメータで元々の最適解を得ることができる．

## 変分不等式

### 相補性問題とKKT条件

$\boldsymbol{F} \colon \mathbb{R}^n \to \mathbb{R}^n$ という関数に対して
\begin{align*}
  \boldsymbol{x} \geq \boldsymbol{0},\quad \boldsymbol{F}(\boldsymbol{x}) \geq \boldsymbol{0},\quad \langle \boldsymbol{F}(\boldsymbol{x}), \boldsymbol{x} \rangle = 0
\end{align*}
を満たす $\boldsymbol{x} \in \mathbb{R}^n$ を求める問題を**相補性問題**という．  
つまり
\begin{align*}
  x_j \geq 0 \quad (j=1,\ldots,n), \quad F_j(x_1,\ldots,x_n) \geq 0 \quad (j=1,\ldots,n), F_j(x_1,\ldots,x_n) x_j = 0 \quad (j=1,\ldots,n)
\end{align*}
を満たす $(x_1,\ldots,x_n) \in \mathbb{R}^n$ を求める問題である．  
最後の等式が$n$個の相補性条件となっているので相補性問題と呼ばれる．

特に $\boldsymbol{F}$ が
\begin{align*}
  \boldsymbol{F}(\boldsymbol{x}) := M \boldsymbol{x} + \boldsymbol{q} \quad (M \in \mathbb{R}^{n \times n}, \boldsymbol{q} \in \mathbb{R}^{n})
\end{align*}
と表せられるとき，相補性問題は線形相補性問題と呼ばれる．

**どうでもいいことかもしれないが**  
$\boldsymbol{q} \geq \boldsymbol{0}$ ならば，$\boldsymbol{x}=\boldsymbol{0}$ は明らかに線形相補性問題の解のひとつ

関数 $\boldsymbol{F}$ が**単調**であるとは，
\begin{align*}
  \langle
  \boldsymbol{F}(\boldsymbol{x}) - \boldsymbol{F}(\boldsymbol{x}),
  \boldsymbol{x} - \boldsymbol{y}
  \rangle
  \geq 0
  \quad (\boldsymbol{x}, \boldsymbol{y} \in \mathbb{R}^{n})
\end{align*}
を満たすことをいう．  
単調な関数に関する相補性問題を単調相補性問題という．

$\boldsymbol{F}_1 \colon \mathbb{R}^n \times \mathbb{R}^m \to \mathbb{R}^n$ と
$\boldsymbol{F}_2 \colon \mathbb{R}^n \times \mathbb{R}^m \to \mathbb{R}^m$ という関数に対して
\begin{align*}
  \boldsymbol{x} \geq \boldsymbol{0}_n,\quad
  \boldsymbol{F}_1(\boldsymbol{x}, \boldsymbol{y}) \geq \boldsymbol{0}_n,\quad
  \langle \boldsymbol{F}_1(\boldsymbol{x}, \boldsymbol{y}), \boldsymbol{x} \rangle = 0 = \boldsymbol{0}_1,\quad
  \boldsymbol{F}_2(\boldsymbol{x},\boldsymbol{y}) = \boldsymbol{0}_m
\end{align*}
を満たす $(\boldsymbol{x}, \boldsymbol{y}) \in \mathbb{R}^n \times \mathbb{R}^m$ を求める問題を**混合相補性問題**という．  
混乱が起きないように $j$ 次元のゼロベクトルを $\boldsymbol{0}_j$ で表したが，以降は省略する．

KKT条件は混合相補性問題である．

image.png

$\boldsymbol{F}_1 \colon \mathbb{R}^m \times \mathbb{R}^n \times \mathbb{R}^r \to \mathbb{R}^m$ と
$\boldsymbol{F}_2 \colon \mathbb{R}^m \times \mathbb{R}^n \times \mathbb{R}^r \to \mathbb{R}^{n \times r}$ を
\begin{align*}
  &\boldsymbol{F}_1(\boldsymbol{\lambda}, \boldsymbol{x}, \boldsymbol{\mu})
  := (-g_1(\boldsymbol{x}),\ldots, -g_m(\boldsymbol{x})))^\top\\
  &\boldsymbol{F}_2(\boldsymbol{\lambda}, \boldsymbol{x}, \boldsymbol{\mu})
  := \left(\nabla f(\boldsymbol{x}) + J \boldsymbol{g}(\boldsymbol{x})^\top \boldsymbol{\lambda} + J \boldsymbol{h}(\boldsymbol{x})^\top \boldsymbol{\mu},
  h_1(\boldsymbol{x}),\ldots, h_r(\boldsymbol{x})\right)^\top
\end{align*}
とおくと
\begin{align*}
  &\boldsymbol{\lambda} \geq \boldsymbol{0}
  \Leftrightarrow
  \lambda_i \geq 0 \quad (i=1,\ldots,m) \\
  &\boldsymbol{F}_1(\boldsymbol{\lambda}, \boldsymbol{x}, \boldsymbol{\mu}) \geq \boldsymbol{0}
  \Leftrightarrow
  g_i(\boldsymbol{x}) \leq 0 \quad (i=1,\ldots,m) \\
  &\langle \boldsymbol{F}_1(\boldsymbol{\lambda}, \boldsymbol{x}, \boldsymbol{\mu}), \boldsymbol{\lambda} \rangle = 0
  \Leftrightarrow
  - \lambda_i g_i(\boldsymbol{x}) = 0 \quad (i=1,\ldots,m)
  \Leftrightarrow
  \lambda_i g_i(\boldsymbol{x}) = 0 \quad (i=1,\ldots,m) \\
  &\boldsymbol{F}_2(\boldsymbol{\lambda}, \boldsymbol{x}, \boldsymbol{\mu}) = \boldsymbol{0}
  \Leftrightarrow
  \nabla f(\boldsymbol{x}) + J \boldsymbol{g}(\boldsymbol{x})^\top \boldsymbol{\lambda} + J \boldsymbol{h}(\boldsymbol{x})^\top \boldsymbol{\mu} = \boldsymbol{0},
  h_1(\boldsymbol{x}) = 0,\ldots, h_r(\boldsymbol{x}) = 0
\end{align*}
となるので，確かにKKT条件は，(2.60b)が本質的に解きたい問題で，制約として(2.60a)と(2.60c)をもつ，混合相補性問題として定式化できる．

### 変分不等式問題

Given a Banach space $E$, a subset $S$ of $E$, and a functional $F \colon S \to E^*$ from $S$ to the dual space $E^*$ of the space $E$, the variational inequality problem is the problem of solving for the variable $x \in S$ the following inequality:
\begin{align*}
\langle F(x),y-x\rangle \geq 0\qquad \forall y\in S \tag{2.91}
\end{align*}
where $\langle \cdot ,\cdot \rangle \colon E^* \times E \to \mathbb{R}$ is the duality pairing.

いまの場合，バナッハ空間（というよりヒルベルト空間）としてユークリッド内積を持つ $\mathbb{R}^n$ を考えているので，双対空間も $\mathbb{R}^n$ となるし，duality pairing はユークリッド内積となるので，あまり深く考えなくてよい．

変分不等式問題は非線形方程式問題の一般化である．  
実際，$S = \mathbb{R}^n$ とすると変分不等式は任意の $y \in S$ に対して満たされなければならないので $y = x - F(x)$ に対しても満たさなければならない．  
するとこのとき変分不等式は $||F(x)|| = 0$ と同値であり，これは $F(x)=0$ と同値である．  
よって $S= \mathbb{R}^n$ の場合の変分不等式問題は「$F(x)=0$ を満たす $x \in \mathbb{R}^n$ を求めよ」という非線形方程式問題となる．

**ちなみに**  
変分問題=変分法を用いた問題  
変分法=ある問題を汎関数の極致問題として捉えて解析する手法  
たとえば「測地線の方程式問題=曲線のエネルギー汎関数の最小値問題」がある（オイラー・ラグランジュ方程式を用いたニュートン力学の解法のひとつ）

image.png

**証明**  
$\boldsymbol{x}$ が (2.87) を満たすとする．  
つまり
\begin{align*}
  &\boldsymbol{x} \geq \boldsymbol{0} \\
  &\boldsymbol{F}(\boldsymbol{x}) \geq \boldsymbol{0} \\
  &\langle \boldsymbol{F}(\boldsymbol{x}), \boldsymbol{x} \rangle = 0
\end{align*}
を満たすとする．  
よって $\boldsymbol{x} \in S$ は明らか．  
任意の $\boldsymbol{y} \in S$ をとると
\begin{align*}
  \langle \boldsymbol{F}(\boldsymbol{x}), \boldsymbol{y} - \boldsymbol{x} \rangle
  = \langle \boldsymbol{F}(\boldsymbol{x}), \boldsymbol{y} \rangle
  = \sum_{i=1}^n F_i(\boldsymbol{x}) y_i
\end{align*}
であるが，$\boldsymbol{F}(\boldsymbol{x}) \geq \boldsymbol{0}$ と $\boldsymbol{y} \geq \boldsymbol{0}$ より，(上式右辺） $\geq 0$ である．  
したがって $\boldsymbol{x}$ は (2.91) の解である．

よって変分不等式問題は相補性問題の拡張である．
