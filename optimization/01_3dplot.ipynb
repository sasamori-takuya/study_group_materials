# np.meshgrid の挙動について確認（メモ）

```
import numpy as np
```

```
x = np.array([0, 1, 2, 3])
y = np.array([4, 5, 6])

print("*** meshgrid(x, y)***")
print("*mesh1*")
mesh1, mesh2 = np.meshgrid(x, y)
print(mesh1.shape)
display(mesh1)
print("*mesh2*")
print(mesh2.shape)
display(mesh2)

print("*** meshgrid(y, x)***")
mesh1, mesh2 = np.meshgrid(y, x)
print("*mesh1*")
print(mesh1.shape)
display(mesh1)
print("*mesh2*")
print(mesh2.shape)
display(mesh2)
```

# 教科書 p26 の図2.1

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import sys

print(sys.version)
print(np.__version__)
```

```
def func1(x, y):
  return x**2 + y**2

def func2(x, y):
  return - x**2 - y**2

N = 50
x1 = np.linspace(-5, 5, N)
x2 = np.linspace(-5, 5, N)
X1, X2 = np.meshgrid(x1, x2)
f1 = func1(X1, X2)
f2 = func2(X1, X2)

fig = plt.figure(figsize=(18,6))

axL = fig.add_subplot(1, 2, 1, projection = "3d")
axL.set_xlabel("x")
axL.set_ylabel("y")
axL.set_zlabel("f1(x, y)")
axL.plot_wireframe(X1, X2, f1)

axR = fig.add_subplot(1, 2, 2, projection = "3d")
axR.set_xlabel("x")
axR.set_ylabel("y")
axR.set_zlabel("f2(x, y)")
axR.plot_wireframe(X1, X2, f2)

plt.show()
```

# 教科書 p27 の例2.2

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import sys

print(sys.version)
print(np.__version__)
```

```
def func(x, y):
  return 10 * 2 + x**2 - 10 * np.cos(np.pi * x / 2) + y**2 - 10 * np.cos(np.pi * y / 2)

N = 50
x1 = np.linspace(-5, 5, N)
x2 = np.linspace(-5, 5, N)
X1, X2 = np.meshgrid(x1, x2)
f = func(X1, X2)
#print(X1.shape)  # (50, 50)
#print(X2.shape)  # (50, 50)
#print(f.shape)  # (50, 50)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.plot_wireframe(X1, X2, f)
plt.show()
```
